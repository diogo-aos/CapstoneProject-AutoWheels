```python
im_list = crops
out_fn = 'crops.zip'

temp_dir = tempfile.mkdtemp()
all_fn = []
z = zipfile.ZipFile(out_fn, 'w') 
for i, im_msg in enumerate(im_list):
    im = bridge.imgmsg_to_cv2(im_msg, im_msg.encoding)
    fn = os.path.join(temp_dir, '{}.jpg'.format(i))
    all_fn.append(fn)
    plt.imsave(fn, im)
z.close()

# clean temporary files
shutil.rmtree(temp_dir)
```
